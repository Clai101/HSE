```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w_o = 3
k = 1  
```

```python
def J_w(w_v):
    return w_v**4 / ((w_o**2 - w_v**2)**2 + w_v**2 * k**2)

w_v = np.linspace(0, 10, 1000)

to__ = 1
max_w_v = np.sqrt(2)*w_o**2/np.sqrt(2*w_o**2 - k**2)
max_value = J_w(max_w_v)
```

```python
fig, ax = plt.subplots(facecolor=(244/255, 228/255, 215/255))
ax.plot(w_v, J_w(w_v), c='black', label = "$\\mathfrak{I} (w_v)$")
ax.plot([max_w_v, max_w_v], [0, max_value], '--', color = 'black')
ax.axhline(y = 1, linestyle = '--', color = 'black')
ax.text(max_w_v, -0.13 * max_value, '$\\frac{\\sqrt{2}w_o^2}{\\sqrt{2w_o^2 - k^2}}$', ha='center')
ax.text(-0.8, to__, '$\\xi$', ha='center')
ax.set_facecolor(color=(244/255, 228/255, 215/255))
ax.grid(True)
plt.xlabel("$w_v$")
plt.ylabel("$\\mathfrak{I}$")
ax.legend(facecolor=(244/255, 228/255, 215/255))
plt.savefig("sours_img/omega.jpg")
```

```python
def J_t(theta):
    return np.sin(theta)**2
```

```python
theta = np.linspace(0, np.pi, 1000)

fig, ax = plt.subplots(facecolor=(244/255, 228/255, 215/255))
ax.plot(theta, J_t(theta), c='black', label = "$\\mathfrak{I} (\\varphi)$")
ax.set_facecolor(color=(244/255, 228/255, 215/255))
ax.grid(True)
plt.xlabel("$\\varphi$")
plt.ylabel("$\\mathfrak{I}$")
ax.legend(facecolor=(244/255, 228/255, 215/255))
plt.savefig("sours_img/phi.jpg")
```
