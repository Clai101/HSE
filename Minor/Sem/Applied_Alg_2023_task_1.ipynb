#### Задание 5
Решить СЛАУ
$$
\left\{
\begin{matrix}
-2.1x + 3y + 4.5z = 10.23\\
3x + 5y + 1.5z = 20.65\\
7.2x + z = 9.34
\end{matrix}
\right.
$$
Для решения СЛАУ воспользоваться linalg.solve, аргументы - матрица левой части и столбец правой.

Для проверки правильности решения использовать allclose.

```python
L = np.array([[-2.1, 3, 4.5], 
             [3, 5, 1.5], 
             [7.2, 0, 1]])
psi = np.array([10.23, 20.65, 9.34]).reshape((3,1))
x = np.linalg.solve(L, psi)
print("x =", x)
print(np.allclose(L @ x, psi))
```
